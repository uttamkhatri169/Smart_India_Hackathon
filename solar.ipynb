```python
import matplotlib.pyplot as plt
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
```

```python
df['Date']
```

```python
solar_df = df['Generation(MU)']
```

```python
solar_df 
```

```python
plt.figure(figsize=(20, 10))
plt.plot(df['Date'].values, solar_df .values, marker='o', linestyle='-', color='g', label='Solar Generation (MU)')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Solar Energy Generation (MU)')
plt.title('Solar Energy Generation Over Time')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
