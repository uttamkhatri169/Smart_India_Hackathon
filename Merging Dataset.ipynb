```python
import pandas as pd

df1 = pd.read_csv('delhi_weather_data.csv')
df2 = pd.read_csv('delhidata.csv')


print("Initial Info of the 1st Dataset:")
df1.info()

print("Initial Info of the 2nd Dataset:")
df2.info()
```

```python
df1.describe()
```

```python
df2.describe()
```

```python
# Merging 'date' and 'time' into a single 'dateTime'
df1['dateTime'] = pd.to_datetime(df1['date'] + ' ' + df1['time'])

# Drop the original 'date' and 'time' 
df1.drop(['date', 'time'], axis=1, inplace=True)


print("Preprocessed Info of the Dataset:")
df1.info()
```

```python
# Merging 'Date' and 'TIMESLOT'
df2['dateTime'] = pd.to_datetime(df2['Date'] + ' ' + df2['TIMESLOT'], format='%d-%m-%Y %H:%M:%S')

# Droping the original 'Date' and 'TIMESLOT'
df2.drop(['Date', 'TIMESLOT'], axis=1, inplace=True)

print("Preprocessed Info of the Dataset:")
df2.info()
```

```python
# Step 2: Merge the two datasets on the 'dateTime' column
mergeddf = pd.merge(df1, df2, on='dateTime', how='inner')
```
